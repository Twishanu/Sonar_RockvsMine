```python
#importing dependencies
import numpy as np
import pandas as pd
#train test split
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Data collection and Data Processing
sonar_data = pd.read_csv('/content/sonar.csv', header=None)
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()#count, mean... of all the columns
```

```python
sonar_data[60].value_counts()#60th column has R or M written
```

```python
sonar_data.groupby(60).mean()
```

```python
#seperating data and labels
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(Y)
```

```python
#Training and test data
#X_train is training data X_test is testing data
#Y_train is training label Y_test is testing label
#X and Y both will be splitted
#test size represents 10% test data 90% training data
#stratify=Y --> equal number of rocks and mines in the training data
#random_state is used to order the split in a specific manner
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
# training our model....Logistic Regression model will be used
model = LogisticRegression()
model.fit(X_train, Y_train)#in the graph one side is X and the other is Y
```

```python
#checking accuracy on training data
X_train_prediction = model.predict(X_train)#in the graph one side is X and the other is Y
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data: ",training_data_accuracy)
```

```python
#checking accuracy on test data....real accuracy
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on test data: ",test_data_accuracy)
```
